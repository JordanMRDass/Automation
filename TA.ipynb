```python
from finta import TA

def symbol_data(Symbols, years_back):
    import pandas as pd
    import datetime
    import yfinance as yf
    
    end = datetime.datetime.now()
    start = datetime.datetime.now() - datetime.timedelta(days=365 * int(years_back))
    # create empty dataframe
    stock_final = pd.DataFrame()
    try:
        # download the stock price 
        stock = []
        stock = yf.download(Symbols,start=start, end=end, progress=False)

        # append the individual stock prices 
        if len(stock) == 0:
            None
        else:
            stock['Name']=Symbols
            stock_final = stock_final.append(stock,sort=False)
    except Exception:
        None
        
    return stock_final

symbol = "AAPL"
years_back = 4
ohlcv_df = symbol_data(symbol, years_back)
```

- Simple moving averages (4 and 100 days)
- Exponential moving averages (5 and 12 days)
- BBANDS
- Relative Strength Index (14 days)
- Directional Movement Index
- Volume-weighted average price
- Pivot Point
- Sentiment analysis

```python
TA_indicators_list = ["SMA", "SMA", "EMA", "EMA", "BBANDS", "RSI", "DMI", "VWAP", "PIVOT_FIB"]
TA_name_list = ["SMA_4", "SMA_100", "EMA_5", "EMA_12", "BBANDS", "RSI_14", "DMI", "VWAP", "PIVOT_FIB"]

list_of_series = [ohlcv_df]
for indicator, name in zip(TA_indicators_list, TA_name_list):
    if len(name.split("_")) == 1 or name.split("_")[-1] == "FIB":
        
        if name in ["BBANDS", "DMI", "PIVOT_FIB"]:
            string_to_eval = f"TA.{indicator}(ohlcv_df)"
        else:
            string_to_eval = f"TA.{indicator}(ohlcv_df).rename('{name}')"
        
    else:
        days = name.split("_")[-1]
        string_to_eval = f"TA.{indicator}(ohlcv_df, {days}).rename('{name}')"
        
    series = eval(string_to_eval)
    list_of_series.append(series)
```

```python
import pandas as pd
import hvplot.pandas

full_df = pd.concat(list_of_series, axis = 'columns', join = 'inner')
```
